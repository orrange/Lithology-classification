```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
import numpy as np
from sklearn.model_selection import validation_curve

from sklearn.svm import SVC
```

```python
#导入数据
dataset= pd.read_excel('data6041.xlsx', sheet_name='Sheet1')
X = dataset.iloc[:, 1:4]
y = dataset.iloc[:,4]
#划分训练集合测试集
X_train,X_test,y_train,y_test = train_test_split (X,y,test_size=0.20,random_state=0)
#标准化、模型训练串联
pipe_lr=Pipeline([('scl',StandardScaler()),('clf',SVC(random_state=0))])
 
```

```python
#case2：验证曲线
param_range=[0.1,1.0,10.0,100,1000,10000,100000]
#10折，验证正则化参数C
train_scores,test_scores =validation_curve(estimator=pipe_lr,X=X_train,y=y_train,param_name='clf__C',param_range=param_range,cv=10)
#统计结果
train_mean= np.mean(train_scores,axis=1)
train_std = np.std(train_scores,axis=1)
test_mean =np.mean(test_scores,axis=1)
test_std=np.std(test_scores,axis=1)
plt.plot(param_range,train_mean,color='blue',marker='o',markersize=5,label='training accuracy')
plt.fill_between(param_range,train_mean+train_std,train_mean-train_std,alpha=0.15,color='blue')
plt.plot(param_range,test_mean,color='green',linestyle='--',marker='s',markersize=5,label='cross_validation accuracy')
plt.fill_between(param_range,test_mean+test_std,test_mean-test_std,alpha=0.15,color='green')
plt.grid()
plt.xscale('log')
plt.xlabel('Parameter C')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim([0,1.0])
plt.savefig('../Data/SVM_C.png', format='png')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import MinMaxScaler


params = [{'solver': 'sgd', 'learning_rate': 'adaptive', 'momentum': 0,
           'learning_rate_init': 0.01},
          {'solver': 'sgd', 'learning_rate': 'adaptive', 'momentum': .9,
           'learning_rate_init': 0.01},
          {'solver': 'sgd', 'learning_rate': 'constant', 'momentum': 0,
           'learning_rate_init': 0.01},
          {'solver': 'sgd', 'learning_rate': 'constant', 'momentum': .9,
            'learning_rate_init': 0.01},
          {'solver': 'sgd', 'learning_rate': 'invscaling', 'momentum': 0,
           'learning_rate_init': 0.01},
          {'solver': 'sgd', 'learning_rate': 'invscaling', 'momentum': .9,
            'learning_rate_init': 0.01},
          {'solver': 'adam', 'learning_rate_init': 0.01}]

labels = ["adaptive learning-rate without momentum",
          "adaptive learning-rate with momentum",
          "constant learning-rate without momentum",
          "constant learning-rate with momentum",
          "invscaling learning-rate without momentum",
          "invscaling learning-rate with momentum",
          "adam"]

plot_args = [{'c': 'red', 'linestyle': '--','linewidth':3},
             {'c': 'red', 'linestyle': '-','linewidth':3},
             {'c': 'blue', 'linestyle': '--'},
             {'c': 'blue', 'linestyle': '-'},
             {'c': 'green', 'linestyle': '--'},
             {'c': 'green', 'linestyle': '-'},
             {'c': 'black', 'linestyle': '-'}]
```

```python
def plot_on_dataset(X, y, ax, name):
    # for each dataset, plot learning for each learning strategy
    print("\nlearning on dataset %s" % name)
    ax.set_title(name)
    X = MinMaxScaler().fit_transform(X)
    mlps = []
    max_iter = 400

    for label, param in zip(labels, params):
        print("training: %s" % label)
        mlp = MLPClassifier(verbose=0, random_state=0,
                            max_iter=max_iter, **param)
        mlp.fit(X, y)
        mlps.append(mlp)
        print("Training set score: %f" % mlp.score(X, y))
        print("Training set loss: %f" % mlp.loss_)
    for mlp, label, args in zip(mlps, labels, plot_args):
            ax.plot(mlp.loss_curve_, label=label, **args)

    
```

```python
fig, axes = plt.subplots(2,1, figsize=(15, 10))
# load / generate some toy datasets
datasets= pd.read_excel('data6041.xlsx', sheet_name='Sheet1')
X = datasets.iloc[:, 1:4]
y = datasets.iloc[:,4]
# iris = datasets.load_iris()
data_sets = [(X, y)]
for ax, data, name in zip(axes.ravel(), data_sets, ['ANN']):
    plot_on_dataset(*data, ax=ax, name=name)
plt.xlabel('training loss')
plt.ylabel('Iterations')
plt.ylim(0,20)
fig.legend(ax.get_lines(), labels=labels, ncol=3, loc="upper center")
plt.savefig('../Data/ANN_adam.png', format='png')
plt.show()
```
